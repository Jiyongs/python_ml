```python
# function
def get_dust_df(path, local):
    dust_df = pd.read_excel(path)
    dust_df2 = dust_df[['측정일시', '측정소명','미세먼지(㎍/㎥)']]
    dust_df2.columns = ['date', 'local', 'dust']
    dust_gu = dust_df2[dust_df2['local']==local]
    dust_gu.index = pd.to_datetime(dust_gu['date'].apply(str))
    
    return dust_gu
```

```python
local = '마포구'
path_tem = r'C:\Users\cypark\Desktop\실습자료\GBS Python 자료\판다스 연습용 데이터\미세먼지\일별평균대기오염도_{}.xlsx'

dust_list = []
for y in range(2010, 2019):
    dust_list.append(get_dust_df(path_tem.format(y), local))

dust_list[1]
```

```python
def get_gradu_df(path):
    grad_df = pd.read_excel(path, sheet_name=1)

    grad_df2 = grad_df[ ['지역', '정보공시 \n 학교코드', '학교명', 
              '졸업자.2', '(특수목적고)과학고 진학자.2',  '(특수목적고)외고ㆍ국제고 진학자.2'] ] 
    grad_df2.columns = ['local', 'code', 'sch_name', 'grad_num', 'sci_num', 'frg_num']
    grad_df3 = grad_df2.drop(0)

    grad_df4 = grad_df3.dropna()
    grad_df4['grad_num'] = pd.to_numeric(grad_df4['grad_num'])
    grad_df4['sci_num'] = pd.to_numeric(grad_df4['sci_num'])
    grad_df4['frg_num'] = pd.to_numeric(grad_df4['frg_num'])

    grad_df5 = grad_df4.reset_index(drop=True)
    
    grad_df5['sp_num'] = grad_df5['sci_num'] + grad_df5['frg_num']
    grad_df5['sp_rate'] = grad_df5['sp_num'] / grad_df5['grad_num'] * 100
    
    return grad_df5
```
