### 데이터 정리
- 컬럼 => 인덱스 => 데이터 값 순서로 정리한다.
- **컬럼** 
    1. 전체 컬럼 중 사용할 컬럼만 추출
    <code>
    df.columns : 컬럼 목록 조회
    df2 = df[ ['col1', 'col2'] ] : 컬럼 선택 
    </code>
    2. 컬럼명을 용이하게 변경
    <code>
    drop() : 컬럼 제거
    </code>
- 엑셀에 셀 병합이 되어 있으면 데이터프레임이 복잡해진다. 원본과 비교하여 필요한 컬럼이 무엇인지 판단해야 한다.
- 데이터프레임의 여러 컬럼을 선택할 때, 인덱스로 컬럼 리스트를 할당한다
    
    
- **인덱스**
    1. 인덱스가 번호일 때, 0번부터 순서대로 나열되어 있는지 확인
    <code>
    df.reset_index(drop=True) : 0부터 시작하는 순차적인 인덱스로 리셋 (drop=True는 기존 인덱스를 버리겠다는 의미)
    </code>
    2. 시계열 데이터일 때, 인덱스가 시간데이터여야 편함
    
    
- **데이터 값**
    1. 결측데이터 = 데이터가 빠져있음 = 'NaN'
    <code>
    df.isna() : NaN 이면 true(1) OR false(0)
    df.isna().sum() : 컬럼별 NaN 값 개수의 합 출력
    df.isna().sum() / len(df) * 100 : 결측 비율 확인
    df.dropna() : 결측데이터(NaN)가 하나라도 있는 행을 제거
    df.loc[ df[컬럼]>90, 컬럼 ] = 변경할 데이터 : dataframe의 컬럼 값을 일괄적으로 변경하는 방법
    </code>
    2. 각 컬럼별 데이터타입
    <code>
    df.dtypes : 컬럼별 데이터타입 조회
    int(df['컬럼']) : str => int
    pd.to_numeric(df['컬럼']) : 데이터 타입을 숫자로 변환
    pd.to_datetime(df['컬럼']) : 데이터 타입을 datetime 으로 변경 (시리즈의 값이 문자열이어야 함)
    </code>

### 함수화

```python
def get_gradu_df(path):
    grad_df = pd.read_excel(path, sheet_name=1)

    grad_df2 = grad_df[ ['지역', '정보공시 \n 학교코드', '학교명', 
              '졸업자.2', '(특수목적고)과학고 진학자.2',  '(특수목적고)외고ㆍ국제고 진학자.2'] ] 
    grad_df2.columns = ['local', 'code', 'sch_name', 'grad_num', 'sci_num', 'frg_num']
    grad_df3 = grad_df2.drop(0)

    grad_df4 = grad_df3.dropna()
    grad_df4['grad_num'] = pd.to_numeric(grad_df4['grad_num'])
    grad_df4['sci_num'] = pd.to_numeric(grad_df4['sci_num'])
    grad_df4['frg_num'] = pd.to_numeric(grad_df4['frg_num'])

    grad_df5 = grad_df4.reset_index(drop=True)
    
    grad_df5['sp_num'] = grad_df5['sci_num'] + grad_df5['frg_num']
    grad_df5['sp_rate'] = grad_df5['sp_num'] / grad_df5['grad_num'] * 100
    
    return grad_df5
```

```python
# 함수화

def get_dust_df(path, local):
    dust_df = pd.read_excel(path)
    dust_df2 = dust_df[ ['측정일시', '측정소명', '미세먼지(㎍/㎥)'] ]
    dust_df2.columns = ['date', 'local', 'dust']
    dust_local = dust_df2[dust_df2['local'] == local]

    dust_local.index = pd.to_datetime(dust_local['date'].apply(str))
    
    return dust_local
```
